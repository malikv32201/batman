```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Y = np.arange(-4, 4,.005)
X = np.zeros((0))

for y in Y:
    X = np.append(X,abs(y/2)-0.009137*y**2+math.sqrt(1-(abs(abs(y)-2)-1)**2)-3)

X = X - 1.25
plt.plot(Y,X,'bo')
plt.show()
```

```python
Y1 = np.append(np.arange(-7, -3, .01), np.arange(3, 7, .01))
X1 = np.zeros((0))
for y in Y1:
    X1 = np.append(X1, 3*math.sqrt(-(y/7)**2+1))
```

```python
X = np.append(X, X1)
Y = np.append(Y,Y1)

plt.plot(Y,X,'bo')
plt.show()
```
